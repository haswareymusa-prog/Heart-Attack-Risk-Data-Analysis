## Dataset

The dataset used in this project is the Heart Attack Risk Prediction Dataset obtained from Kaggle. It contains demographic, health, and lifestyle information for 8,763 patient records.

# Basic Heart Attack Risk Data Analysis

This project uses Python to examine basic information from a heart attack risk dataset. The analysis summarizes the size of the dataset, patient age and cholesterol values, and the number and percentage of patients in each heart attack risk category.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart_attack_prediction_dataset.csv")
df.head()
```

```python
print("Number of patients:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
print(df.columns)
```

```python
print("Average age:", round(df["Age"].mean(), 2))
print("Youngest age:", df["Age"].min())
print("Oldest age:", df["Age"].max())

print()

print("Average cholesterol:", round(df["Cholesterol"].mean(), 2))
print("Lowest cholesterol:", df["Cholesterol"].min())
print("Highest cholesterol:", df["Cholesterol"].max())
```

```python
df["Heart Attack Risk"].value_counts()
```

```python
risk_percentages = df["Heart Attack Risk"].value_counts(normalize=True) * 100
print(risk_percentages.round(2))
```

```python
df["Heart Attack Risk"].value_counts().sort_index().plot(kind="bar")

plt.title("Heart Attack Risk Groups")
plt.xlabel("Heart Attack Risk")
plt.ylabel("Number of Patients")
plt.xticks([0, 1], ["No Risk", "Risk"], rotation=0)
plt.show()
```

## Conclusion

The dataset contains 8,763 patient records and 26 columns. The patients range from 18 to 90 years old, with an average age of approximately 54 years. The average cholesterol level was approximately 260, with values ranging from 120 to 400.

About 64% of the patients were labeled without heart attack risk, while approximately 36% were labeled with heart attack risk. This analysis provides a basic description of the dataset but does not determine whether age or cholesterol causes heart attack risk.
